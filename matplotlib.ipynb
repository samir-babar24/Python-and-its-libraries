```python
#matplotlib 2:markers(used to denote points by various shape)
import matplotlib.pyplot as plt
import numpy as np
names=np.array(["sammir","radhey","atul"])
cgpa=np.array([7.8,8,9.8])
plt.plot(names,cgpa,marker='o')
plt.show()
#markers are=o,*,1,2,3 and so on

```

```python
#matplotlib 3:
#color and the line formaat
import matplotlib.pyplot as plt
import numpy as np
names=np.array(["sammir","radhey","atul"])
cgpa=np.array([7.8,8,9.8])
plt.plot(names,cgpa,'o:b')
plt.show()
#here o works as marker
#here ':' works as type of line
#here b= color of the line
```

```python
#matplotlib 4: giving names to x,y,main axis
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([60,67,70,55])
plt.plot(x,y,'*--r')
title="class and their strengths"
xlab="divison"
ylab="strength"
plt.title(title)
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.show()
```

```python
#matplotlib 5:various operation on markers
import matplotlib.pyplot as plt
import numpy as np
names=np.array(["sammir","radhey","atul","laxman","raju"])
cgpa=np.array([9,7,6,9,8])
plt.plot(names,cgpa,marker='o',ms=20,mec='r',mfc='violet')
plt.title("students gpa")
plt.xlabel("name")
plt.ylabel("gpa")
plt.show()
#here ms=maraker size
#mec=marker edge color
#mfc=marker face color
```

```python
#matplotlib 7double line
x=np.array([3,4,6,7])
y=np.array([4,8,5,1])
plt.plot(x,'1-r',ms=10,mec='y',mfc='violet',linewidth=10)
plt.plot(y,'2:b',ms=15,mec='b',mfc='#4CAF50',linewidth=15)
plt.show()
#here '#4caf50' is a green color
```

```python
#matplotlib 8: title,xlable,ylables various operations
import matplotlib.pyplot as plt
import numpy as np
Building_name=np.array(["mechanical","civil","aiml","computer"])
Building_no=np.array([9,12,5,6])
plt.plot(Building_name,Building_no,'*--r',ms=10,mec='y',mfc='#4CAF50')
#here i am now creating fonts for adding in lables
#family refers to style

font1={'family':'sherif','color':'blue','size':15}
font2 ={'family':'sherif','color':'red','size':20}
font3={'family':'sherif','color':'violet','size':10}
plt.title("building name and no",fontdict=font1)
plt.xlabel("building name ",fontdict=font2)
plt.ylabel("building  no",fontdict=font3)#herefontdict used to call font
plt.show()


```

```python
#matplotlib 9:method 2 for giving fontadict in simple way
import matplotlib.pyplot as plt
import numpy as np
Building_name=np.array(["mechanical","civil","aiml","computer"])
Building_no=np.array([9,12,5,6])
plt.plot(Building_name,Building_no,'*--r',ms=10,mec='y',mfc='#4CAF50')
plt.title("building name and no",fontdict={'family':'sherif','color':'blue','size':15},loc='left')#here loc=position of the title
plt.xlabel("building name ",fontdict={'family':'sherif','color':'red','size':20})
plt.ylabel("building  no",fontdict={'family':'sherif','color':'violet','size':10})#herefontdict used to call font
plt.show()
#without making an seperate font section we have directly operated this
```

```python
#matplotlib 12: subplot
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
name= np.array(["s", "trupti", "yash", "tanmy"])
prn= np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)#1=row,2=column,1=at which no. we want to have our graph
plt.plot(name,prn)
plt.title("div b")


#plot 2:
name1 = np.array(["sahil", "sangam", "rahul", "keshav"])
prn2= np.array([1, 20, 3, 4])

plt.subplot(1, 2, 2)#1=roow,2=column,2=place
plt.plot(name1,prn2)
plt.title("div a")
plt.suptitle("students data")#suptitle used for main title in subplots
plt.show()
```

```python
#matplotlib 13: bars
import matplotlib.pyplot as plt
import numpy as np
pc_games=np.array(["farcry3","callofdutyAW","callofdutyMw","farcry5","gta5"])
rating_out_of10=np.array([7,5,8,4,9])
plt.bar(pc_games,rating_out_of10)
plt.show()
```

```python
#matplotlib 14: operations in bar
import matplotlib.pyplot as plt
import numpy as np
pc_games=np.array(["farcry3","callofdutyAW","callofdutyMw","farcry5","gta5"])
rating_out_of10=np.array([7,5,8,4,9])
plt.bar(pc_games,rating_out_of10,color='red',width=0.25)
plt.show()
```

```python
#matplotlib 16: some operation on pie
import matplotlib.pyplot as plt
import numpy as np
l=["ram","sham","harish","samir"]
marks=np.array([20,30,10,40])
plt.pie(marks,labels=l,startangle=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
l=["ram","sham","harish","samir"]
marks=np.array([20,30,10,40])
e=[0.2,0.3,0.1,0]
myc=["r","b","hotpink","#4CAF50"]
plt.pie(marks,labels=l,startangle=90,explode=e,shadow=True,colors=myc)#here exploding in parts , shadow meaning shadow undow exploded part
#color is udes to give colors
plt.show()
```
